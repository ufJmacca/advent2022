```python
import requests
import tomli

with open("secrets.toml", mode="rb") as fp:
    config = tomli.load(fp)

r = requests.get('https://adventofcode.com/2022/day/1/input', headers={'cookie': config['day-1']['cookie']})
```

```python
calories = 0
highest_calories = 0
calories_list = []

for line in r.text.splitlines():
    if line == '':
        calories_list.append(calories)
        if highest_calories <= calories:
            highest_calories = calories
        calories = 0
        continue
    calories += int(line)


print(highest_calories)    

print(sum(sorted(calories_list, reverse=True)[0:3]))
```
